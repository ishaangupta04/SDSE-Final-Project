## 4) Create a new column 'Education_Bin'
+  'Education_Bin' must account for all degree types and store them in unique numerical values
+   i.e. Bachelor's = 0, Master's = 1, PhD = 2, etc. 

```python
education = []
for value in df.loc[:, 'Education Level']:
    if value == "Bachelor's":
        education.append(0)
    elif value == "Master's":
        education.append(1)
    elif value == "PhD":
        education.append(2)
    else:
        education.append(5)

df['Education_Bin'] = education
```

## 5) Create a new column 'Gender_Bin'
+  'Gender_Bin' must account for all genders within the dataset and store them in binary classified values
+    i.e. Female = 0, Male = 1

```python
gender = []
for value in df.loc[:, 'Gender']:
    if value == "Female":
        gender.append(0)
    elif value == "Male":
        gender.append(1)
    else:
        gender.append(2)

df['Gender_Bin'] = gender
```

## 6) Remove any missing (NaN) values from the dataset
+ Use the .dropna() function to remove any missing values from the dataset. We don't need them!
+ Print the final, cleaned dataset

```python
df = df.dropna(how = 'any')
df
```
